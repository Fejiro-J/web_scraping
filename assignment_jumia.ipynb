```python
import requests
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.jumia.com.ng/led-lcd-tvs/'
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent':userAgent}
response = requests.get(url, headers=headers)

response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
sauce = soup.find_all('div', class_ = 'info')
sauce
```

```python
for i in sauce:
    print(i.text)
```

```python
names = []
prices = []
oldprices = []
no_of_ratings = []
```

```python
for i in sauce:
    if i.find('h3', class_ = 'name') is not None:
        name = i.find('h3', class_ = 'name')
        names.append(name.text)
    else:
        names.append('none')

    #price
    if i.find('div', class_ = "prc") is not None:
        price = i.find('div', class_ = "prc")
        prices.append(price.text)
    else:
        prices.append('none')

    #oldprice
    if i.find('div', class_ = 's-prc-w') is not None:
        oldprice = i.find('div', class_ = 's-prc-w')
        oldprices.append(oldprice.text)
    else:
        oldprices.append('none')

    #no of rating
    if i.find('div', class_ = 'rev') is not None:
        ratings = i.find('div', class_ = 'rev')
        no_of_ratings.append(ratings.text)
    else:
        no_of_ratings.append('none')


info_df = pd.DataFrame({ 'names':names, 'prices':prices, 'oldprices':oldprices, 'no_of_ratings':no_of_ratings})

info_df
```

```python
info_df['old_prices'] = info_df['oldprices'].str.replace('%','')
info_df['old_prices']
```

```python
info_df[['star_rating', 'valid_ratings']] = info_df['no_of_ratings'].str.split('(', expand = True)
```

```python
info_df['valid_ratings'] =  info_df['valid_ratings'].str.replace(')','')

info_df
```
